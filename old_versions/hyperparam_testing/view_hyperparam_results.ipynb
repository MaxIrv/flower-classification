```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('hyperparam_test_results.csv')

# Convert string representations of lists back to lists
df['train_accuracies'] = df['train_accuracies'].apply(ast.literal_eval)
df['train_losses'] = df['train_losses'].apply(ast.literal_eval)
df['val_accuracies'] = df['val_accuracies'].apply(ast.literal_eval)
df['val_losses'] = df['val_losses'].apply(ast.literal_eval)
```

```python
df
```

```python
# Define the values
learning_rate_values = [0.005, 0.001, 0.0005, 0.0001, 0.00005, 0.00001, 0.000005]
learning_rate_default = 0.0001

batch_size_values = [8, 16, 32, 64, 128, 256]
batch_size_default = 64

weight_decay_values = [0.1, 0.01, 0.001, 0.0001, 0.00001]
weight_decay_default = 0.01

df['learning_rate'].unique()

df.drop(columns=['Unnamed: 0'], inplace=True)
# Drop rows where learning_rate = 0.005 because it was a mistake
df = df[df.learning_rate != 0.005]
```

```python
def plot_losses(df, param):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
    
    for i, row in df.iterrows():
        ax1.plot(row['train_losses'], label=f'train {row[param]}')
        ax2.plot(row['val_losses'], label=f'val {row[param]}')
    
    ax1.set_title(f'Training Losses for {param}')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.legend()
    
    ax2.set_title(f'Validation Losses for {param}')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Loss')
    ax2.legend()
    
    plt.tight_layout()
    plt.show()
    
def plot_accuracies(df, param):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
    
    for i, row in df.iterrows():
        ax1.plot(row['train_accuracies'], label=f'train {row[param]}')
        ax2.plot(row['val_accuracies'], label=f'val {row[param]}')
    
    ax1.set_title(f'Training Accuracies for {param}')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Accuracy')
    ax1.legend()
    
    ax2.set_title(f'Validation Accuracies for {param}')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy')
    ax2.legend()
    
    plt.tight_layout()
    plt.show()
```

```python
learning_rate = df.loc[(df['batch_size'] == batch_size_default) & (df['weight_decay'] == weight_decay_default)]
plot_losses(learning_rate, 'learning_rate')
plot_accuracies(learning_rate, 'learning_rate')
```

```python
batch_size = df.loc[(df['learning_rate'] == learning_rate_default) & (df['weight_decay'] == weight_decay_default)]
plot_losses(batch_size, 'batch_size')
plot_accuracies(batch_size, 'batch_size')
```

```python
weight_decay = df.loc[(df['learning_rate'] == learning_rate_default) & (df['batch_size'] == batch_size_default)]
plot_losses(weight_decay, 'weight_decay')
plot_accuracies(weight_decay, 'weight_decay')
```

```python
# Get the number of epochs for each row
df['num_epochs'] = df['train_accuracies'].apply(len)
```
