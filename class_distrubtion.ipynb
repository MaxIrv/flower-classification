# Visualise the Class Distribution in Train/Test/Validation splits

### Imports

```python
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
```

### Setup Datasets

```python
# Define the transformations
transformations1 = transforms.Compose([transforms.ToTensor(), transforms.Resize((250, 250))])

# Load the dataset
training_dataset = torchvision.datasets.Flowers102(root='./data', split="train", 
                                                   download=True, transform=transformations1)
testing_dataset = torchvision.datasets.Flowers102(root='./data', split="test",
                                                   download=True, transform=transformations1)
validation_dataset = torchvision.datasets.Flowers102(root='./data', split="val",
                                                   download=True, transform=transformations1)
```

### Count Classes Function

```python
def count_classes(dataset):
  class_counts = [0] * 102  # There are 102 classes in the Flowers-102 dataset
  for _, label in dataset:
      class_counts[label] += 1
  return class_counts
```

### Count classes in each Dataset

```python
# Count classes in each dataset
train_class_counts = count_classes(training_dataset)
test_class_counts = count_classes(testing_dataset)
val_class_counts = count_classes(validation_dataset)
```

### Plot the class distributions

```python
# Plot the class distributions
def plot_class_distribution(class_counts, dataset_type):
    plt.figure(figsize=(15, 5))
    plt.bar(range(102), class_counts)
    plt.xlabel('Class Label')
    plt.ylabel('Number of Instances')
    plt.title(f'Class Distribution in {dataset_type} Dataset')
    plt.show()

# Plot distributions
plot_class_distribution(train_class_counts, 'Training')
plot_class_distribution(test_class_counts, 'Testing')
plot_class_distribution(val_class_counts, 'Validation')
```
